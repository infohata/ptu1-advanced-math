```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
mpg = sns.load_dataset('mpg')
mpg.info()
```

```python

mpg[mpg['horsepower'].isnull()]
```

```python
mpg['horsepower'].loc[32] = 75
mpg['horsepower'].loc[126] = 105
mpg['horsepower'].loc[330] = 51
mpg['horsepower'].loc[336] = 120
mpg['horsepower'].loc[354] = 73
mpg['horsepower'].loc[374] = 82
mpg.info()
```

```python
origins = pd.get_dummies(mpg['origin'])
origins[100:115]
```

```python
data = pd.concat([mpg, origins], axis=1)
data.tail()
```

```python
data.drop(['origin', 'name'], axis=1, inplace=True)
```

```python
data[111:117]
```

```python
x = data.drop('mpg', axis=1)
x.tail()
```

```python
y = data['mpg']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=91)
# x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(x, y)
# reg
```

```python
sns.regplot(x='displacement', y='mpg', data=data)
```

```python
prediction = reg.predict(x)
prediction
```

```python
spejimai = pd.Series(data=prediction, name='Spejimas')
spejimai
```

```python
rezultatas = pd.concat([y.reset_index(), spejimai], axis=1)
rezultatas
```

```python
reg.score(x, y)
```

```python
rezultatas['klaida'] = rezultatas['mpg'] - rezultatas['Spejimas']
rezultatas
```

```python
sns.histplot(rezultatas['klaida'])
```

```python
su_pavadinimais = pd.concat([mpg['name'], rezultatas.set_index('index')], axis=1)
su_pavadinimais[su_pavadinimais['Spejimas'].notna()]
```
