```python
import pandas as pd
import numpy as np

df = pd.read_csv('miestai.csv')
df
```

```python
df.set_index('Miestas', inplace=True)
```

```python
for metai in df:
    print(metai)
```

```python
df
```

```python
df['mazeja'] = True
for miestas in df.index:
    gyventoju = df.loc[miestas, '2019'].copy()
    for metai in df.columns:
        if metai.isnumeric() and int(metai) < 2019:
            dabar = df.loc[miestas, metai].copy()
            if gyventoju >= dabar:
                df.loc[miestas, 'mazeja'] = False
                break
            gyventoju = dabar.copy()
# DONE
```
