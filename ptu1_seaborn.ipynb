```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

# %matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
sns.histplot(tips['total_bill'], bins=50)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='size', size='size', sizes=(20, 200), style='time', legend='full', markers={'Dinner': 'o', 'Lunch': 'X'})
```

```python
tips.tail(20)
```

```python
sns.pairplot(tips, hue='sex', diag_kind='hist')
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, hue='day', estimator=sum, palette=sns.husl_palette(4, h=.7, l=.8))
```

```python
sns.countplot(x='time', data=tips)
```

```python
sns.despine()
sns.boxplot(x='sex', y='tip', data=tips, palette=sns.husl_palette(2, h=.2))
```

```python
tips.describe()
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
sns.set_style('darkgrid')
gridux = sns.FacetGrid(data=tips, col='time', row='sex')
gridux.map(sns.scatterplot, 'tip', 'total_bill')
```
